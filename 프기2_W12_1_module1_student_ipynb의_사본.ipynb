## 3.1. main2.py 함수를 만든다. 이 함수에서 calculations.py 를 임포트하여 각 계산을 수행하도록 한다. 이 때, calculations 모듈을 cal 로 줄여서 부르도록 import 해보자.

````python
import calculations as cal # cal 로 줄여서 불러보자.

# 1 + 2
print('add:1 + 2:', cal.add(1,2)) # 모듈 cal 에서 add 함수로 계산


# 1 - 2
print('sub:1 - 2:', cal.sub(1,2)) # 모듈 cal 에서 sub 함수로 계산


# 2 * 3
print('multi: 2*3 ', cal.add(2,3)) # 모듈 cal 에서 multi 함수로 계산



# 3/2
print('divide: 3 / 2:', cal.add(3,2)) # 모듈 cal 에서 divide 함수로 계산

````

## 3.2. main3.py 에서는 calculations 모듈의 특정 함수를 함수 이름만 쓸 수 있는 형태로 불러와보자.

````python
# from 모듈이름 import 모듈함수

from calculations import add


# 1 + 2
print('add:1 + 2:', add(1,2)) # 모듈 cal 에서 add 함수로 계산


````
